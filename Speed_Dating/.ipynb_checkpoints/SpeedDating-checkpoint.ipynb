# Speed Dating

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('max_columns', None)
```

```python
# Setting options for displaying dataset
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
pd.set_option('display.width', 1000)
```

```python
# Reading dataset and printing 5 first rows
dataset = pd.read_csv('Speed_Dating_Data.csv', encoding = 'Latin')

dataset.head()
```

# Data Exploration

```python
#Returning Shape of dataset
print("Number of rows and columns: {}".format(dataset.shape))
print()

#  Calculating the proportion of missing values in each feature
print("Percentage of missing values: ")
display(100*dataset.isnull().sum()/dataset.shape[0])
```

There are many missing values, more than 50% in many cases, and that effects our results. Therefore, we will just disregard these fields with over 50% of null values from the dataset and narrow our analysis to the filed that we can use.

```python
# Age distribution of participants
sns.distplot(dataset['age'], kde=False, color = 'blue', bins=int(180/5),  # binwidth
             hist_kws={'edgecolor':'black'})
plt.xlabel('Age')
plt.ylabel('Density')
```

From the plot above we see that most of the participants are between 20 and 30.

```python
# Replacing 0 and 1 with female and male respectively
gender = dataset.gender.apply(lambda x: "female" if x==0 else "male")

# Countplot for gender
sns.set(rc={"figure.figsize": (3,6)})
sns.countplot(gender, palette = 'Set2')
```

The number of male and female participants are almost the same. 

```python
race = dataset['race'].apply(lambda x: 'Black/African American' if x == 1 
                         else 'European/Caucasian-American' if x == 2
                         else 'Latino/Hispanic American' if x == 3
                         else 'Asian/Pacific Islander/Asian-American' if x == 4
                         else 'Native American' if x == 5
                         else 'Other')

sns.set(rc={"figure.figsize": (6,6)})
sns.countplot(y =race, hue = gender, palette = 'Set1')
```

From the plot above we observe that the participants are mainly European/Caucasian-American and there is not any Native American. 

```python
### Match when Same Race
same_race = dataset.loc[dataset['samerace'] == 1, ['match']]
same_race = same_race.match.apply(lambda x: "No" if x == 0 else "Yes")

### Match when Not Same Race
not_same_race = dataset.loc[dataset['samerace'] == 0, ['match']]
not_same_race = not_same_race.match.apply(lambda x: "No" if x == 0 else "Yes")


fig, axes = plt.subplots(1, 2, figsize=(15, 8))

# Same Race
sns.countplot(same_race, ax = axes[0], hue = gender, palette = 'Set2')
axes[0].set_title('Match when same race', fontsize=15)



# Not Same Race
sns.countplot(not_same_race, hue = gender, ax = axes[1], palette = 'Set2')
axes[1].set_title('Match when no same race', fontsize=15)



plt.show()
```

```python
dataset['samerace'].value_counts()
```

There is a preferance in the 'non same race' counting around 5000 participants while in the 'same race' counts around 3000. Although in both cases we do not have many matches.

```python
age = dataset['age_o'] # age of partner
rating = dataset['attr_o'] # rating by partner the night of the event

plt.figure(figsize = (10, 10))
sns.scatterplot(age, rating)
plt.xlabel('Age')
plt.ylabel('Attractiveness Rating')
plt.show()
```

As expected there is a variaty of the overall attractiveness rating versus the age. We do not notice any correlation between these features. 

```python
# Goal of the events for participants

plt.figure(figsize = (6, 6))
goal = dataset.goal.apply(lambda x: 'Seemed like a fun night out' if x == 1
                          else 'To meet new people' if x == 2
                          else 'To get a date' if x == 3
                          else 'Looking for a serious relationship' if x == 4
                          else 'To say I did it' if x == 5
                          else 'Other')

sns.countplot(y = goal, hue = gender, palette = 'Set1')
```

It looks like that people mostly joined those events to have fun and to meet new people.
Very few are looking for a serious relationship.

Now, we will investigate the decision of partner the night of event.

```python
plt.figure(figsize = (6, 6))
dec = dataset.goal.apply(lambda x: 'Yes' if x == 1
                          else 'No')

sns.countplot(dec, hue = gender)
```

## Count by Study Category

```python
def get_field_label(x):
    labels = {
        1: "Law",
        2: "Math",
        3: "Social Science",
        4: "Medical",
        5: "Engineering",
        6: "Journalism",
        7: "History",
        8: "Business",
        9: "Education",
        10: "Biological",
        11: "Social Work",
        12: "Undergrad",
        13: "Political",
        14: "Film",
        15: "Fine Arts",
        16: "Languages",
        17: "Architecture",
        18: "Other",
    }
    return labels.get(x)

dataset["field_label"] = dataset["field_cd"].apply(get_field_label)

fig, ax = plt.subplots(figsize=(20,10))
plt.xticks(rotation=20)
graph = sns.countplot(x = dataset["field_label"], ax = ax)
```

## Match percentage by field study

```python
serieField = dataset.groupby('field_label').count()['iid']
serieMatch = dataset[dataset['match'] == 1].groupby('field_label').count()['iid']

# Percentage of match according to the type of study

percentageYesMatch = serieMatch / serieField * 100
percentageYesMatch

percentageDataFrame = pd.DataFrame({ 'percentage': percentageYesMatch, 'field_label': percentageYesMatch.keys() })
g = sns.catplot(x = "field_label", y = "percentage", kind = 'bar', data = percentageDataFrame, aspect = 3)
g.set_xticklabels(rotation=30)
```

## Correlation between match and total like

```python
dataset['like_total'] = dataset['like'] + dataset['like_o']
like_total = dataset['like_total'].value_counts()

df1 = pd.DataFrame({ 'count': like_total, 'like_total': like_total.keys() }, index = like_total.keys()).sort_values(by = 'like_total')
df2 = dataset[dataset["match"] == 1].groupby(["like_total"]).count()[["iid"]]

total_like_data = df1.merge(df2, on = "like_total")
total_like_data["percentage"] = total_like_data["iid"] / total_like_data["count"] * 100
filtered_total = total_like_data[total_like_data["count"] > 19]

sns.lmplot(x = 'like_total', y = 'percentage', data = filtered_total)
```

As expected the more the total like increase the more the percentage of match augment.
